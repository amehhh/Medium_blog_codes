```python
import matplotlib.pyplot as plt 
import numpy as np 
import os 
from matplotlib.image import imread
plt.rcParams['figure.figsize']=[16,18]

#Import the photo into python
A= imread(r'C:\Users\AMEH EMMANUEL SUNDAY\Downloads\SVD photo.jpg')
X=np.mean(A,-1) ; #Convert the RGB image to grayscale 
print(f"The size of the coloured image is {A.shape}")
print(f"The size of the white and black  image is {X.shape}")

#Plot the original coloured image
plt.figure(figsize=(3, 3))
plt.imshow(A)
plt.axis('off')
plt.show()

#Plot the white and black image version
plt.figure(figsize=(3, 3))
plt.imshow(X,cmap='gray')
plt.axis('off')
plt.show()
```

```python
U,S,VT =np.linalg.svd(X,full_matrices=False) #Full_matrices=False means that we want the Economy Singular Value Decomposition
num_singular_values = len(S)  # This gives r = min(m, n)

#Print the first singular value 
print(f"First singular Value is {S[1]} and last singular value is {S[-1]}")
print(f"The number of singular values is {num_singular_values}")
S=np.diag(S) # This extracts and diagonalizes values from the singular matrix 
j=0 
for r in (5,10,20,100,120):
  #Construct approximate image ( Low rank approximate images)
  Xapprox= U[:,:r] @ S[0:r,:r] @ VT[:r,:]
  #U is an mxm matrix
  #S is an mxn matrix
  #VT is an nvn matrix
  plt.figure(j+1)
  plt.figure(figsize=(3, 3))  # Adjust the size (width x height in inches)
  j+=1
  img=plt.imshow(Xapprox)
  img.set_cmap('gray')
  plt.axis('off')
  plt.title('r=' + str(r))
  plt.show()
```

```python
#PLOTTING
plt.figure(figsize=(6,3))
plt.figure(1)
plt.semilogy(np.diag(S))
plt.title('Singular Values - Fig 1')
plt.xlabel('Singular values j in descending order')
plt.ylabel('Energy (Singular values)')
plt.show()



plt.figure(2)
plt.figure(figsize=(6,3))
plt.plot(np.cumsum(np.diag(S)) / np.sum(np.diag(S)))
plt.xlabel('Singular values j in descending order')
plt.ylabel('R singular values / N singular values')
plt.title('Singular Values: Cummulative Sum - Fig 2')
plt.show()
```
